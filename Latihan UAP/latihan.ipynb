```python
import Bio
from Bio.Seq import Seq
from Bio.SeqUtils import molecular_weight as mw
from Bio.SeqUtils import MeltingTemp as mt
from matplotlib import pyplot as plt
```

```python
seqA = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
seqB = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")
```

```python
print(mw(seqA))
print(mw(seqB))
```

```python
mtA = mt.Tm_Wallace(seqA)
mtB = mt.Tm_Wallace(seqB)

plt.pie([mtA, mtB], labels=['seqA', 'seqB'])
plt.show()
```

```python
def gc_content(seq):
    gc = [x for x in seq if x in 'GC']
    return (len(gc)/len(seq)) * 100

def at_content(seq):
    at = [x for x in seq if x in 'AT']
    return (len(at) / len(seq)) * 100

def gc_content2(seq):
    gc = float(seq.count('G') + seq.count('C')) / len(seq)
    return gc * 100


def at_content2(seq):
    at = float(seq.count('A') + seq.count('T')) / len(seq)
    return at * 100

print(gc_content2(seqA))
print(gc_content(seqA))
print(gc_content2(seqB))
print(at_content2(seqA))
print(at_content2(seqB))
```

```python
plt.bar(['GCA', 'GCB', 'ATA', 'ATB'], [gc_content2(seqA), gc_content2(seqB), at_content2(seqA), at_content2(seqB)])
plt.show()
```

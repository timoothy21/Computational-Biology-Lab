```python
import Bio
from Bio.Seq import Seq
```

```python
# Create Seqeunce
seq1 = Seq('CGATCGATCGAT')
seq2 = Seq('ATATATCGAT')
```

```python
# GC Content
# Manual 1
def gc_content(seq):
    result = float(seq.count('G') + seq.count('C')) / len(seq)
    return result * 100

gc_content(seq1)
```

```python
# Manual 2
def gc_content2(seq):
    gc = [x for x in seq if x in 'GC']
    return (len(gc)/len(seq)) * 100

gc_content2(seq1)
```

```python
# GC Function
from Bio.SeqUtils import GC

print(GC(seq1))
```

```python
# GC fraction
from Bio.SeqUtils import gc_fraction
gc_fraction(seq1) * 100
```

```python
# AT Manual
def at_content(seq):
    result = float(seq.count('A') + seq.count('T')) / len(seq)
    return result * 100

at_content(seq1)
```

```python
# AT Manual 2
def at_content2(seq):
    result = [x for x in seq if x in 'AT']
    return (result / len(seq)) * 100

at_content(seq1)
```

# Melting Point

```python
# Wallace Manual:
def gc_len(seq):
    gc = [x for x in seq if x in 'GC']
    return len(gc)

def at_len(seq):
    at = [x for x in seq if x in 'AT']
    return len(at)

gc = gc_len(seq2)
at = at_len(seq2)

def wallace_manual(seq):
    mt = 2 * at_len(seq) + 4 * gc_len(seq)
    return mt

print(wallace_manual(seq2))
```

```python
# Wallace Function
from Bio.SeqUtils import MeltingTemp as mt

mt.Tm_Wallace(seq2)
```

```python
# GC
print(mt.Tm_GC(seq2))
```

```python
import Bio
from Bio.Seq import Seq
from Bio.SeqUtils import molecular_weight as mw
from Bio.SeqUtils import MeltingTemp as mt
from matplotlib import pyplot as plt
```

```python
seqA = Seq("AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC")
seqB = Seq("CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG")
```

```python
# Determine molecular weight of both sequence

print(mw(seqA))
print(mw(seqB))
```

```python
# Calculate melting point from seq A and B
wall_seqA = mt.Tm_Wallace(seqA)
wall_seqB = mt.Tm_Wallace(seqB)

print(wall_seqA)
print(wall_seqB)
```

```python
# pie chart
data = [wall_seqA, wall_seqB]
label = ["Melting point A", "Melting point B"]
plt.pie(data, labels=label)
plt.show()
```

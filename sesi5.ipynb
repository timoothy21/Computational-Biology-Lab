PDB => Protein Data Bank
DAta yang akan diambil dari PDB
    File:
        - Model = keseluruhan data protein
        - Chain = protein sequence
        - Residue = 1 Codon Amino Acid
        - Atom = atom yang membentuk codon amino acid

FASTA => Menyimpan DNA Sequence
    - Id
    - Description
    - DNA Seuqence

GB => Gen Bank = menyimpan DNA Sequence
    - Id
    - Description
    - DNA Sequence
    - Features

```python
import Bio

from Bio.PDB.PDBParser import PDBParser
```

```python
parser = PDBParser()
structure_pdb = parser.get_structure('6lu7', './6lu7.pdb')
structure_pdb
```

```python
# Model
model = structure_pdb[0]

for chain in model:
    print(chain)
    for residue in chain:
        print(residue)
        for atom in residue:
            print(atom)
```

```python
from Bio import SeqIO
```

```python
fasta_record = SeqIO.parse('./coronavirus.fasta', 'fasta')
for record in fasta_record:
    print(record.id)
    print(record.description)
    print(record.seq)
    for protein in record.seq.translate().split('*'):
        print(protein)
        print(len(protein))
```

```python
fasta_record = SeqIO.read('./coronavirus.fasta', 'fasta')
print(fasta_record.id)
print(fasta_record.description)
print(fasta_record.seq)
```

```python
genbank_record = SeqIO.read('./coronavirus.gb', 'genbank')
print(genbank_record.id)
print(genbank_record.description)
print(genbank_record.seq)
for feature in genbank_record.features:
    print(feature)

for protein in genbank_record.seq.translate().split('*'):
    print(protein)
```

```python
seqA = genbank_record.seq[:round(len(genbank_record.seq)/3)]
print(len(seqA))

seqB = genbank_record.seq[round(len(genbank_record.seq)/3):round(len(genbank_record.seq)*2/3+1)]
print(len(seqB))
```

```python
from Bio.SeqUtils import MeltingTemp as mt

mt_seqA = mt.Tm_NN(seqA)
mt_seqB = mt.Tm_NN(seqB)

print(mt_seqA, mt_seqB)
```

```python
from matplotlib import pyplot as plt

plt.bar(['seqA', 'SeqB'], [mt_seqA, mt_seqB])
plt.show()
```
